## E 2)

If you think about the task of spam classification more thoroughly, do you think that the classification accuracy or misclassification error is a good error metric of how good an email classifier is? What are potential pitfalls? (Hint: think about false positives [non-spam email classified as spam] and false negatives [spam email classified as non-spam]).

Relying solely on accuracy or misclassification error to evaluate an email spam classifier can be misleading due to several potential pitfalls. One key issue is the imbalanced nature of the dataset, where the majority of emails are non-spam. In such cases, a model could achieve high accuracy simply by classifying most emails as non-spam, but this would fail to effectively detect spam emails. Additionally, false positives, where non-spam emails are mistakenly classified as spam, can lead to users missing important messages, causing frustration or even significant consequences. Conversely, false negatives, where spam emails are misclassified as non-spam, allow harmful or unwanted emails to reach the inbox, posing security risks and cluttering the user's mailbox.

To address these issues, better evaluation metrics should be considered. Precision focuses on the proportion of correctly identified spam emails out of all emails classified as spam, reducing the risk of false positives. Recall measures the proportion of actual spam emails that are correctly identified, minimizing false negatives. The F1 score provides a balance between precision and recall, making it particularly useful for imbalanced datasets. Additionally, the ROC-AUC metric assesses the model's overall ability to discriminate between spam and non-spam emails across various thresholds, offering a more comprehensive evaluation of performance.

In the k-nearest neighbor (k-NN) algorithm, no computation happens during the training phase, as it simply stores the entire training dataset. ## E 20)

Consider the following code cell, where I implemented *k*-nearest neighbor classification algorithm following the the scikit-learn API

````python
import numpy as np


class KNNClassifier(object):
    def __init__(self, k, dist_fn=None):
        self.k = k
        if dist_fn is None:
            self.dist_fn = self._euclidean_dist

    def _euclidean_dist(self, a, b):
        dist = 0.
        for ele_i, ele_j in zip(a, b):
            dist += ((ele_i - ele_j)**2)
        dist = dist**0.5
        return dist

    def _find_nearest(self, x):
        dist_idx_pairs = []
        for j in range(self.dataset_.shape[0]):
            d = self.dist_fn(x, self.dataset_[j])
            dist_idx_pairs.append((d, j))

        sorted_dist_idx_pairs = sorted(dist_idx_pairs)

        return sorted_dist_idx_pairs

    def fit(self, X, y):
        self.dataset_ = X.copy()
        self.labels_ = y.copy()
        self.possible_labels_ = np.unique(y)

    def predict(self, X):
        predictions = np.zeros(X.shape[0], dtype=int)
        for i in range(X.shape[0]):
            k_nearest = self._find_nearest(X[i])[:self.k]
            indices = [entry[1] for entry in k_nearest]
            k_labels = self.labels_[indices]
            counts = np.bincount(k_labels,
                                 minlength=self.possible_labels_.shape[0])
            pred_label = np.argmax(counts)
            predictions[i] = pred_label
        return predictions
````
